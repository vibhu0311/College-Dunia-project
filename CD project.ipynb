```python
#cleaning data by formatting to the datetime datatype
df['Instance Date'] = pd.to_datetime(df['Instance Date'], errors='coerce')
df['Last Lead Activity Date'] = pd.to_datetime(df['Last Lead Activity Date'], errors='coerce')

df['Last Lead Activity Date'].unique().tolist()
```

```python
#function for getting avg days in between instance day and last lead activity
def datediff(d):
    s=0
    for i in d:
        if i.days>0:
            s+= i.days
    return s/len(d)
```

```python
#average days from first instance to last day of lead activity for different types of leads
print("Time taken for all leads")
print(datediff(df['Last Lead Activity Date']-df['Instance Date']),"\n")

print("Application Submitted")
g=df.groupby(['Lead Stage'])
grp=g.get_group('Application Submitted')
print (datediff(grp['Last Lead Activity Date']-grp['Instance Date']),"\n")

print("Hot")
g=df.groupby(['Lead Stage'])
grp=g.get_group('Hot')
print (datediff(grp['Last Lead Activity Date']-grp['Instance Date']),"\n")

print("Warm")
g=df.groupby(['Lead Stage'])
grp=g.get_group('Warm')
print (datediff(grp['Last Lead Activity Date']-grp['Instance Date']),"\n")

print("Cold")
g=df.groupby(['Lead Stage'])
grp=g.get_group('Cold')
print (datediff(grp['Last Lead Activity Date']-grp['Instance Date']),"\n")
```
